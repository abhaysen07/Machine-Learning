```python

#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
data  =  datasets.load_iris()    #iris data
```

```python
#print(data.DESCR)
data
print(data.target_names)
```

```python
#split data

x   = data.data[:, [2, 3] ]  

#data.target_names

#data.feature_names
y  =  data.target
```

```python
#import algorithm

from sklearn.tree import DecisionTreeClassifier
```

```python
#initialize model

model  = DecisionTreeClassifier(random_state= 1 , max_depth= 2)
```

```python
model
```

```python
#train and test split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=100)

X_train.shape , X_test.shape
```

```python
#train model
model.fit(X_train , y_train)
```

```python
predictions  =  model.predict(X_test)
predictions
```

```python
x   = data.data 

#data.target_names

#data.feature_names
y  =  data.target

y
```

```python
#import random forest algorithm

from sklearn.ensemble import RandomForestClassifier

#tensorflow implemantations   
```

```python
model  = RandomForestClassifier(random_state=10 , max_depth= 2 , n_estimators =5) 
```

```python
model
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=100)
```

```python
#train model 

model.fit(X_train, y_train)
```

```python
predictions =  model.predict(X_test) #predict on  test data
```

```python
predictions
```

```python
#test visualizations 

from sklearn import tree
```

```python
iris_feature_names =  data.feature_names
iris_class_names  =  data.target_names

fig, axes  =  plt.subplots(nrows=1 , ncols=5 , figsize  =  (10,2), dpi = 900)

for index in range(0,5):
    
    tree.plot_tree(model.estimators_[index ],
                    feature_names=iris_feature_names,
                    class_names=iris_class_names,
                    filled= True, 
                    ax =  axes[index])
    axes[index].set_title('estimator:' + str(index), fontsize  = 10)
    
    
fig.savefig('random_forest.png')
    
```
